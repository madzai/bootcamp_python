```python
def sum(x, y):
  return x + y

print(sum(7, 9))
print(sum('abc', 'def'))
```

```python
def operation(x, y):
  return x + y, x - y, x * y, x / y

r1, r2, r3, r4 = operation(10, 4)
print(f"r1={r1}, r2={r2}, r3={r3}, r4={r4}")
```

```python
def _2dp(num):
  return round(num, 2)

x = 9.2345
print(_2dp(x))
```

```python
# hint for parameter & return type
def is_odd(x: int) -> bool:
  return x % 2 == 1

print(is_odd(10))
print(is_odd(1))
print(is_odd(10.5)) # False
# print(is_odd('John')) # error
```

```python
# Type Change
names = ['Leo', 'Lucas', 'Tommy']
name_set = set(names)
name_list = list(names)

print(type(names))
print(names)
print(type(name_set))
print(name_set)
print(type(name_list))
print(name_list)
```

```python
# Function (Lambda)
square_formula = lambda x: x ** 2
print(square_formula(8))

# Java: Stream
numbers = [7, 10, 100, 4, 5]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)

names = ['John', 'Tom', 'Jennie', 'Lucas']
# len > 3
names3 = list(filter(lambda x: len(x) > 3, names))
print(names3)
```
