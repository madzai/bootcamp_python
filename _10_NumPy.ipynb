```python
import numpy as np

arr = np.array([2, 5, 10])
print(arr)
print(type(arr))

arr2d = np.array([[2, 3], [9, 10], [4, 7], [10, 12]])
print(arr2d)
print(type(arr2d))

print(arr + 1)
print(arr * 2)

print(np.sum(arr))
print(np.median(arr))
print(np.mean(arr))
print(np.std(arr))
print(np.var(arr))
print(np.min(arr))
print(np.max(arr))

print(np.sort(arr))
print(np.unique(arr)) # similar to set

# list
list = [2, 3, 6]
print(list * 2)
```
