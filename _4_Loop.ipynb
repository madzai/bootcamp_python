```python
# Basic Math Method
numbers = [1, 10, 6, 100, -9]
# Static methods
print(max(numbers))
print(min(numbers))
print(sum(numbers))
print(len(numbers))

# Find index by number
print(numbers.index(max(numbers)))
print(numbers.index(min(numbers)))

# Average
print(sum(numbers) / len(numbers))
```
