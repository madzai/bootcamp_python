```python
import json

json_str = '{"name" : "John", "age" : 17}'
dict1 = json.loads(json_str)
print(dict1)
print(type(dict1))

print(dict1["name"])
print(dict1["age"])

json_str_missing_comma = '{"name" : "John" "age" : 17}'
# dict2 = json.loads(json_str_missing_comma) # error

# from dict to json
json_str2 = json.dumps(dict1, indent=2)
print(json_str2)
```

```python
with open("user1.json") as file:
  user_dict = json.load(file)
  print(user_dict)
```

```python
cat_str = '{"name" : "Mary", "gender" : "F"}'
cat_dict = json.loads(cat_str)
with open("cat.json", "w") as file:
  json.dump(cat_dict, file, indent=2)
```

```python
with open("persons.json", "r") as file:
  person_list = json.load(file)

print(person_list)
print(type(person_list))

# lambda
# "USA"
persons_usa = list(filter(lambda x: x['city'] == 'USA', person_list))
print(persons_usa)

persons_china = list(filter(lambda x: x['address']['country'] == 'China', person_list))
print(persons_china)
```
