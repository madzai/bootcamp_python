```python
import math

print(math.sqrt(9))
print(math.sqrt(9.5))
# print(math.sqrt(-9.5))

print(math.pow(3, 4))
print(math.pow(3, 4.5))

print(math.ceil(3.3))
print(math.floor(3.9))

print(math.pi)

radius = 3.5
area = radius ** 2 * math.pi
print(area)

print(3 ** 3)
```

```python
import statistics

numbers = [20, 10, 90, 99, 40, 24, 90]
print(statistics.median(numbers))
print(statistics.mean(numbers))
print(statistics.stdev(numbers))
print(statistics.mode(numbers))
```
