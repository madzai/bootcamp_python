```python
# String + int -> error
# s = 'hello' + 3
s = 'hello' + str(3) # explicit conversion
print(s)

# print('goodbye' + False)
print('goodbye' + str(False))
```

```python
# array, String
word = 'Python'

print(word[0])
print(word[-1]) # Java: length - 1
print(word[-2])

print(word[:3])
print(word[0:3])

print(word[3:])
print(word[2:3])

# str
print(word.upper())
print(word.lower())
print(word.replace('y', 'a'))
print(word)
```

```python
# str method
studentname = 'Steven'
print(studentname.count('e'))
print(studentname.isdigit()) # False
print(studentname.isalpha()) # True
print(studentname.startswith('s')) # False
print(studentname.startswith('S')) # True, case sensitive

score = '93'
print(score.isdigit()) # True
```
